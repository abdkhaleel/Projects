```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for column in df.columns:
    if df[column].dtype == object:
        df[column] = le.fit_transform(df[column])
```

```python
df.head()
```

```python
y = df['Recurred']
X = df.drop(['Recurred'], axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
GBC = RandomForestClassifier(criterion='log_loss', n_estimators=50)
```

```python
GBC.fit(X_train, y_train)
```

```python
y_pred = GBC.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
GBC_scaled = RandomForestClassifier(criterion='log_loss', n_estimators=50)
```

```python
GBC_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred_scaled = GBC_scaled.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred_scaled)
```

```python
f1_score(y_test, y_pred_scaled)
```

```python
confusion_matrix(y_test, y_pred_scaled)
```
